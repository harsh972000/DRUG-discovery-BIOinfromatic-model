```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
pd.set_option('display.max_columns', 500)
```

recovered the data from ChEMBL data base

```python
target = new_client.target
target_q = target.search('Acetylcholinesterase')
targets = pd.DataFrame.from_dict(target_q)
targets
```

```python
sele_tar = targets.target_chembl_id[0]
sele_tar
```

Data is filter by IC50 value

```python
activeity = new_client.activity
res  = activeity.filter(target_chembl_id = sele_tar).filter(standard_type= 'IC50')
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
dff = df[['canonical_smiles', 'molecule_chembl_id', 'standard_value']].dropna(how = 'any')
```

ligand data is now classified as inactive or active on basis of IC50 value 

```python
new =[]
for i in dff['standard_value']:
    if float(i) >= 10000:
        new.append('inactive')
    elif float(i) <=1000:
        new.append('active')
    else: new.append('intermediate')
```

```python
dff['class'] = new
```

```python
df_f = dff[['canonical_smiles', 'molecule_chembl_id']]
```

making a smile file of the data 

```python
df_f.to_csv('mole_build.smi', sep='\t', index=False, header=False)
```

```python
fp = dict(zip(FP_list, xml_files))
fp
```

by using pedelpy library, the molecular driscriptor of ligand has been made 

```python
from padelpy import padeldescriptor
fingerprint = 'Substructure'
fingerprint_output_file = ''.join([fingerprint,'.csv'])
fingerprint_descriptortypes = fp[fingerprint]

padeldescriptor(mol_dir='mole_build.smi', d_file=fingerprint_output_file, 
                descriptortypes= fingerprint_descriptortypes,
                detectaromaticity=True,
                standardizenitro=True,
                standardizetautomers=True,
                threads=2,
                removesalt=True,
                log=True,
                fingerprints=True)

descriptors = pd.read_csv(fingerprint_output_file)
```

```python
x = descriptors.drop(['Name'], axis =1 )
```

```python
y  = dff['class']
```

now the machine learning model RandomForestClassifier is build and data is fed to the ML model for training.
the data is devided into training and testing set for crossvalidation and check the score

```python
!pip install sklearn
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=3, train_size=0.80)
from sklearn.ensemble import RandomForestClassifier
import numpy as np
np.random.seed(32)
randomf  = RandomForestClassifier()
randomf.fit(x_train, y_train)
```
